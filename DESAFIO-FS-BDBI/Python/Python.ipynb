```python
#Plotando um grafico de barras que exiba a quantidade de funcionários em cada setor
df2 = df.groupby('Department')[['EmployeeCount']].count()
```

```python
df2.plot (kind = 'bar',color ='blue',edgecolor = 'black',figsize = (10,5))

plt.xlabel('Setor')
plt.ylabel('Quantidade de empregados')
plt.title('Quantidade de empregados  em casa setor')
plt.xticks(rotation = 40)

plt.show()
```

```python
#Plotando gráfico de linha que exiba a quantidade de funcionários para cada nível de escolaridade
df3 = df.groupby('Education')[['EmployeeCount']].count()
```

```python
df3.plot(kind = 'line',color = 'purple',marker = 'o',figsize = (10,6))

plt.xlabel('Nível de educação')
plt.ylabel('Funcionários')
plt.title('Quantidade de empregados por nível de escolaridade')

plt.show()
```

```python
#Plotando um que exiba a quantidade de funcionários que tem ou não algum atrito na empresa
df4 = df.groupby('Attrition')['EmployeeCount'].count()
```

```python
df4.plot(kind ='pie',colors = ['blue','red'], startangle=20 , autopct='%1.1f%%')
plt.title('Quantidade de Funcionarios com atrito ou não')
plt.show()
```

```python
#Plotando um histogram que exibe total de anos por pessoa
sns.histplot(df['TotalWorkingYears'], bins=20, kde = True)
plt.show
```
