```python
#Agrupando os dados da coluna Department e calculando a media de horas trabalhada
df2 = df.groupby('Department')[['TotalWorkingYears']].sum()
df2.plot(kind = 'barh', color = 'red')
plt.title('Soma de Anos de Trabalho por Departamento')
plt.xlabel('Anos de Trabalho')
plt.ylabel('Departamento')
plt.show()
```

```python
#Plotando gráfico que mostra a quantidade de funcionarios por departamento
df3 = df.groupby('Department')['EmployeeCount'].count()
df3.plot(kind = 'bar', color = 'blue')
plt.title('Quantidade de Funionarios por Departamento')
plt.xlabel('Quantidade de Funcionarios')
plt.ylabel('Departamento')
plt.show()
```
